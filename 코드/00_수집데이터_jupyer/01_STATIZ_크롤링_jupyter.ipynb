# Crawling

```python
import os
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# 셀레늄 연결
chromedriver = 'C:/Users/kyu/Downloads/chromedriver_win32/chromedriver.exe'
browser = Chrome(chromedriver)
time.sleep(1)

# 브라우저로 STATIZ 접속하기 
browser.get('http://www.statiz.co.kr/player.php')
time.sleep(1)

# 전처리 테이블 정의
clean_list = []

# 선수데이터 수집
for player in only_player:
    
    try:
        # 검색창 검색어 입력
        search_area = browser.find_element_by_css_selector('#search_text')
        search_area.send_keys(player)
        time.sleep(0.5)

        # 검색하기 클릭
        browser.find_element_by_css_selector('#search-btn').send_keys(Keys.ENTER)
        time.sleep(0.5)   

        # 날짜별 클릭
        browser.find_element_by_css_selector('body > div > div.content-wrapper > div > section.content > div > div:nth-child(1) > div > div.col-xs-12.col-sm-8.col-md-8.col-lg-12 > div > div.table-responsive.no-padding.box > table > tbody > tr > td > a:nth-child(4)').send_keys(Keys.ENTER)
        time.sleep(0.5)
    
    except:
        # 선수가 없는 경우
        continue 

    # 년도 반복문
    idx = 1
    while True:
        try:
            browser.find_element_by_css_selector('body > div > div.content-wrapper > div > section.content > div > div:nth-child(2) > div > div:nth-child(2) > div > a:nth-child(' + str(idx) +')').send_keys(Keys.ENTER)
            time.sleep(0.5)
            
            #년도 저장
            year = browser.find_element_by_css_selector('body > div > div.content-wrapper > div > section.content > div > div:nth-child(2) > div > div:nth-child(2) > div > a:nth-child(' + str(idx) +')').text

            # 테이블 로딩
            root1 = browser.find_elements_by_tag_name('table')[2]

            # 원테이블 저장(raw_list)
            raw_list = [] #원테이블 초기화
            for i in root1.find_elements_by_tag_name('tr'):
                raw_list.append(i.text)

            # 원테이블 전처리(clean_list)
            for i in raw_list:
                if '월' in i:
                    pass
                else:
                    sp = year+' '+player+' '+i
                    clean_list.append(sp.split())

            # 중간보고
            print(year + '년 ' + player + ' 수집완료')

            #다음년도 숫자설정
            idx = idx+1

        except:
            # 결과보고
            print(player + ' 데이터 수집완료')
            print('----------------------------')
            break
```
